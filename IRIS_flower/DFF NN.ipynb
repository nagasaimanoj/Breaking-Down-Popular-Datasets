# IRIS Flower Classification

## Classifying Iris flower types using Deep Feed Forward Neural Network

### Imports

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

from matplotlib.pyplot import scatter, plot

from sklearn.datasets import load_iris
```

### Dataset

```
iris = load_iris()

train_x = iris['data']
train_y = to_categorical(iris['target'])
```

### Model Creation

```
model = Sequential([
    Dense(units=16, input_dim=4),
    Dense(units=32, activation='softmax'),
    Dense(units=3, activation='softmax')
])
model.summary()
```

Training the model

```
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['categorical_accuracy'])
model.fit(train_x, train_y, epochs=100)
prediction = model.predict(iris['data'])
```

### Plotting the results

```
scatter(range(len(iris['data'])), [list(i).index(max(i)) for i in prediction])
scatter(range(len(iris['data'])), iris['target'])
plot()
```
