```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_graphviz
import pandas as pd
import numpy as np
import graphviz
```

```python
iris_df = pd.DataFrame(load_iris()['data'])
```

```python
iris_df['species'] = [load_iris()['target_names'][i] for i in load_iris()['target']]
iris_df.columns = load_iris()['feature_names'] + ['species']
```

```python
iris_df.head()
```

```python
tree = DecisionTreeClassifier().fit(iris_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']], iris_df['species'])
```

```python
dot_data = export_graphviz(tree) 
graph = graphviz.Source(dot_data) 
graph
```

```python
dot_data = export_graphviz(tree, out_file=None, 
                     feature_names=load_iris()['feature_names'],  
                     class_names=load_iris()['target_names'],  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
graph
```
